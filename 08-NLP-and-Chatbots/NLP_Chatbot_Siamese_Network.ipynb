```python
!pip install -q sentence-transformers tensorflow pandas scikit-learn matplotlib

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, Model, Input, optimizers
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
import os
import pickle
import random
import time

# Set seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

# ---- Expanded Q&A Dataset ----
# Original + new related questions
zc_qa_dataset = [
    {"question": "What are the admission requirements for Zewail City?",
     "answer": "Admission to Zewail City requires completing secondary education with high scores, passing the admission test, and an interview. International applicants need equivalent qualifications."},
    {"question": "How can I apply to Zewail City?",
     "answer": "To apply to Zewail City, visit the official website, create an account, fill out the online application form, upload required documents, and pay the application fee."},
    {"question": "What programs does Zewail City offer?",
     "answer": "Zewail City offers undergraduate and graduate programs in Engineering, Science, and Business. Popular programs include Nanotechnology, Biomedical Sciences, Renewable Energy, and Materials Science."},
    {"question": "When is the application deadline for Zewail City?",
     "answer": "The standard application deadline for Zewail City is typically in June for the fall semester. Early applications are encouraged as programs may fill up quickly."},
    {"question": "Is financial aid available at Zewail City?",
     "answer": "Yes, Zewail City offers various financial aid options including scholarships, grants, and work-study programs based on academic merit and financial need."},
    {"question": "What is the tuition fee for Zewail City?",
     "answer": "Tuition fees at Zewail City vary by program but typically range from 70,000 to 120,000 EGP per academic year. Additional fees apply for laboratory courses and materials."},
    {"question": "Does Zewail City offer accommodation for students?",
     "answer": "Yes, Zewail City provides on-campus housing facilities for both male and female students, with priority given to international and out-of-town students."},
    {"question": "What is the minimum GPA required for admission to Zewail City?",
     "answer": "Zewail City typically requires a minimum GPA of 3.2 on a 4.0 scale or 85-90% in secondary education. However, actual requirements may vary by program."},
    {"question": "Are there any entrance exams for Zewail City?",
     "answer": "Yes, Zewail City requires applicants to take an entrance examination that tests knowledge in mathematics, physics, and English proficiency."},
    {"question": "Can international students apply to Zewail City?",
     "answer": "Yes, Zewail City welcomes international students. They must provide equivalency certificates for their secondary education qualifications and may need to demonstrate English proficiency."},
    {"question": "What documents are required for admission?",
     "answer": "Required documents include official transcripts, personal statement, recommendation letters, standardized test scores (if applicable), ID/passport copy, and application fee payment receipt."},
    {"question": "How can I contact the admissions office?",
     "answer": "You can contact Zewail City's admissions office by email at admissions@zewailcity.edu.eg, by phone at +20-123456789, or through the contact form on their official website."},
    {"question": "Does Zewail City offer PhD programs?",
     "answer": "Yes, Zewail City offers PhD programs in various fields including Nanotechnology, Materials Science, Physics, Biomedical Sciences, and other specializations. Each program has specific requirements and research focus areas."},
    {"question": "Where is Zewail City located?",
     "answer": "Zewail City of Science and Technology is located in 6th of October City, on the outskirts of Cairo, Egypt. The campus is designed to provide a modern research and learning environment."},
    {"question": "What facilities are available on campus?",
     "answer": "Zewail City campus features state-of-the-art research laboratories, modern lecture halls, digital library, student center, sports facilities, cafeterias, and residential buildings for student accommodation."},
    {"question": "What prerequisites do I need for Zewail City admission?",
     "answer": "Admission to Zewail City requires completing secondary education with high scores, passing the admission test, and an interview. International applicants need equivalent qualifications."},
    {"question": "How do I submit my application to Zewail City?",
     "answer": "To apply to Zewail City, visit the official website, create an account, fill out the online application form, upload required documents, and pay the application fee."},
    {"question": "What majors can I study at Zewail City?",
     "answer": "Zewail City offers undergraduate and graduate programs in Engineering, Science, and Business. Popular programs include Nanotechnology, Biomedical Sciences, Renewable Energy, and Materials Science."},
    {"question": "When should I apply to Zewail City?",
     "answer": "The standard application deadline for Zewail City is typically in June for the fall semester. Early applications are encouraged as programs may fill up quickly."},
    {"question": "Are scholarships available at Zewail City?",
     "answer": "Yes, Zewail City offers various financial aid options including scholarships, grants, and work-study programs based on academic merit and financial need."},
    {"question": "How much does it cost to study at Zewail City?",
     "answer": "Tuition fees at Zewail City vary by program but typically range from 70,000 to 120,000 EGP per academic year. Additional fees apply for laboratory courses and materials."},
    {"question": "Is there student housing available at Zewail City?",
     "answer": "Yes, Zewail City provides on-campus housing facilities for both male and female students, with priority given to international and out-of-town students."},
    {"question": "What grades do I need to get into Zewail City?",
     "answer": "Zewail City typically requires a minimum GPA of 3.2 on a 4.0 scale or 85-90% in secondary education. However, actual requirements may vary by program."},
    {"question": "Do I need to take an exam to get into Zewail City?",
     "answer": "Yes, Zewail City requires applicants to take an entrance examination that tests knowledge in mathematics, physics, and English proficiency."},
    {"question": "Does Zewail City accept students from other countries?",
     "answer": "Yes, Zewail City welcomes international students. They must provide equivalency certificates for their secondary education qualifications and may need to demonstrate English proficiency."},
    {"question": "What papers do I need to submit for Zewail City application?",
     "answer": "Required documents include official transcripts, personal statement, recommendation letters, standardized test scores (if applicable), ID/passport copy, and application fee payment receipt."},
    {"question": "How do I reach the Zewail City admissions team?",
     "answer": "You can contact Zewail City's admissions office by email at admissions@zewailcity.edu.eg, by phone at +20-123456789, or through the contact form on their official website."},
    {"question": "Can I pursue a doctorate degree at Zewail City?",
     "answer": "Yes, Zewail City offers PhD programs in various fields including Nanotechnology, Materials Science, Physics, Biomedical Sciences, and other specializations. Each program has specific requirements and research focus areas."},
    {"question": "Where exactly is Zewail City situated?",
     "answer": "Zewail City of Science and Technology is located in 6th of October City, on the outskirts of Cairo, Egypt. The campus is designed to provide a modern research and learning environment."},
    {"question": "What amenities does Zewail City campus have?",
     "answer": "Zewail City campus features state-of-the-art research laboratories, modern lecture halls, digital library, student center, sports facilities, cafeterias, and residential buildings for student accommodation."},
]

# Create DataFrame with the dataset
df = pd.DataFrame(zc_qa_dataset)
print(f"Total QA pairs: {len(df)}")

# ---- Data Preparation ----
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove extra whitespace
    text = ' '.join(text.split())
    return text

# Apply preprocessing
df['processed_question'] = df['question'].apply(preprocess_text)
df['processed_answer'] = df['answer'].apply(preprocess_text)

# ---- Create Training Data ----
# 1. Generate positive pairs (same answer)
# 2. Generate negative pairs (different answers)

def create_pairs_dataset(df):
    pairs = []
    labels = []

    # Group by answer to find questions with the same answer
    answer_groups = df.groupby('answer')['processed_question'].apply(list).to_dict()

    # Create positive pairs (questions with same answer)
    for answer, questions in answer_groups.items():
        if len(questions) > 1:  # At least 2 questions needed to form a pair
            for i in range(len(questions)):
                for j in range(i+1, len(questions)):
                    pairs.append((questions[i], questions[j]))
                    labels.append(1)  # Positive pair

    # Create negative pairs (questions with different answers)
    all_answers = list(answer_groups.keys())
    for i in range(len(all_answers)):
        for j in range(i+1, len(all_answers)):
            answer1 = all_answers[i]
            answer2 = all_answers[j]

            # Take one question from each answer
            q1 = answer_groups[answer1][0]
            q2 = answer_groups[answer2][0]

            pairs.append((q1, q2))
            labels.append(0)  # Negative pair

    # Augment with more negative pairs
    if len(labels) - sum(labels) < sum(labels):  # If fewer negative than positive pairs
        needed_negatives = sum(labels) - (len(labels) - sum(labels))

        while needed_negatives > 0:
            # Randomly select two different answers
            answer1, answer2 = random.sample(all_answers, 2)

            # Take random questions from each answer
            q1 = random.choice(answer_groups[answer1])
            q2 = random.choice(answer_groups[answer2])

            pairs.append((q1, q2))
            labels.append(0)
            needed_negatives -= 1

    return np.array(pairs), np.array(labels)

pairs, labels = create_pairs_dataset(df)

print(f"Total pairs: {len(pairs)}")
print(f"Positive pairs: {sum(labels)}")
print(f"Negative pairs: {len(labels) - sum(labels)}")

# Show a few examples
print("\nExample positive pair:")
pos_idx = np.where(labels == 1)[0][0]
print(f"Q1: {pairs[pos_idx][0]}")
print(f"Q2: {pairs[pos_idx][1]}")

print("\nExample negative pair:")
neg_idx = np.where(labels == 0)[0][0]
print(f"Q1: {pairs[neg_idx][0]}")
print(f"Q2: {pairs[neg_idx][1]}")

# ---- Generate Transformer Embeddings ----
print("\nLoading transformer model...")
# Using a multilingual model to better handle potential languages
transformer_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

print("\nGenerating embeddings for all questions...")
# Extract unique questions from the pairs
all_questions = set()
for pair in pairs:
    all_questions.add(pair[0])
    all_questions.add(pair[1])

print(f"Total unique questions: {len(all_questions)}")

# Generate embeddings for all unique questions
question_to_embedding = {}
for question in all_questions:
    embedding = transformer_model.encode(question)
    question_to_embedding[question] = embedding

# Create embeddings numpy array for all pairs
embedding_dim = list(question_to_embedding.values())[0].shape[0]
print(f"Embedding dimension: {embedding_dim}")

X1 = np.zeros((len(pairs), embedding_dim))
X2 = np.zeros((len(pairs), embedding_dim))

for i, (q1, q2) in enumerate(pairs):
    X1[i] = question_to_embedding[q1]
    X2[i] = question_to_embedding[q2]

# ---- Build Siamese Network ----
def euclidean_distance(vectors):
    x, y = vectors
    sum_square = tf.reduce_sum(tf.square(x - y), axis=1, keepdims=True)
    return tf.sqrt(tf.maximum(sum_square, tf.keras.backend.epsilon()))

def cosine_distance(vectors):
    x, y = vectors
    x_norm = tf.nn.l2_normalize(x, axis=1)
    y_norm = tf.nn.l2_normalize(y, axis=1)
    # Cosine distance = 1 - cosine similarity
    return 1.0 - tf.reduce_sum(x_norm * y_norm, axis=1, keepdims=True)

def contrastive_loss(y_true, y_pred):
    margin = 1.0
    square_pred = tf.square(y_pred)
    margin_square = tf.square(tf.maximum(margin - y_pred, 0))
    return tf.reduce_mean(y_true * square_pred + (1 - y_true) * margin_square)

def create_siamese_model(input_shape):
    # Define the layers for the shared network
    input_a = Input(shape=(input_shape,))
    input_b = Input(shape=(input_shape,))

    # Shared network layers
    shared_network = tf.keras.Sequential([
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),
        layers.Dense(64, activation='relu'),
        layers.BatchNormalization(),
        layers.Dense(32, activation='tanh')
    ])

    # Process both inputs through the shared network
    processed_a = shared_network(input_a)
    processed_b = shared_network(input_b)

    # Calculate distance between the outputs
    distance = layers.Lambda(euclidean_distance)([processed_a, processed_b])

    # Create the Siamese model
    siamese_model = Model(inputs=[input_a, input_b], outputs=distance)

    # Create the encoder model for inference (embedding generation)
    encoder_model = Model(inputs=input_a, outputs=processed_a)

    return siamese_model, encoder_model, shared_network

# Create the Siamese model
siamese_model, encoder_model, shared_network = create_siamese_model(embedding_dim)

# Compile the model
siamese_model.compile(
    loss=contrastive_loss,
    optimizer=optimizers.Adam(learning_rate=0.001)
)

siamese_model.summary()

# Split the data into training and validation sets
X1_train, X1_val, X2_train, X2_val, y_train, y_val = train_test_split(
    X1, X2, labels, test_size=0.2, random_state=42, stratify=labels
)

# ---- Train the Siamese Network ----
# Callbacks for training
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6
)

# Train the model
print("\nTraining Siamese network...")
history = siamese_model.fit(
    [X1_train, X2_train],
    y_train,
    batch_size=32,
    epochs=100,
    validation_data=([X1_val, X2_val], y_val),
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
```

```python
# ---- Evaluate the Model ----
# Predictions on the validation set
val_preds = siamese_model.predict([X1_val, X2_val])

# Convert distances to similarity scores (0 to 1)
val_preds_flat = val_preds.flatten()
similarity_scores = 1 - np.clip(val_preds_flat / 2.0, 0, 1)  # Normalize distances to [0,1]

# Find optimal threshold
from sklearn.metrics import precision_recall_curve, f1_score, accuracy_score, roc_curve, roc_auc_score

precision, recall, thresholds = precision_recall_curve(y_val, similarity_scores)
f1_scores = 2 * precision * recall / (precision + recall + 1e-6)
optimal_threshold_idx = np.argmax(f1_scores)
optimal_threshold = thresholds[optimal_threshold_idx]

# Predictions using optimal threshold
binary_preds = (similarity_scores >= optimal_threshold).astype(int)
accuracy = accuracy_score(y_val, binary_preds)
f1 = f1_score(y_val, binary_preds)

print(f"\nModel Evaluation:")
print(f"Optimal threshold: {optimal_threshold:.4f}")
print(f"Validation Accuracy: {accuracy:.4f}")
print(f"Validation F1 Score: {f1:.4f}")

# Plot ROC curve
fpr, tpr, _ = roc_curve(y_val, similarity_scores)
auc = roc_auc_score(y_val, similarity_scores)

plt.subplot(1, 2, 2)
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], 'k--')
plt.title(f'ROC Curve (AUC = {auc:.4f})')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.tight_layout()
plt.savefig('siamese_training_results.png')
plt.show()

```

```python
# ---- Test with Example Queries ----
# Function to find the most similar question
def find_most_similar_question(query, questions_df, transformer_model, encoder_model, threshold=optimal_threshold):
    # Preprocess the query
    processed_query = preprocess_text(query)

    # Get transformer embedding for query
    query_transformer_embedding = transformer_model.encode([processed_query])[0]

    # Get Siamese embedding for query
    query_siamese_embedding = encoder_model.predict(np.array([query_transformer_embedding]), verbose=0)[0]

    # Calculate similarity with all questions
    all_questions = questions_df['processed_question'].tolist()
    similarities = []

    for question in all_questions:
        # Get pre-calculated embeddings
        question_transformer_embedding = question_to_embedding[question]

        # Get Siamese embedding
        question_siamese_embedding = encoder_model.predict(np.array([question_transformer_embedding]), verbose=0)[0]

        # Calculate cosine similarity between Siamese embeddings
        similarity = np.dot(query_siamese_embedding, question_siamese_embedding) / \
                    (np.linalg.norm(query_siamese_embedding) * np.linalg.norm(question_siamese_embedding))

        similarities.append(similarity)

    # Find the most similar question
    max_similarity_idx = np.argmax(similarities)
    max_similarity = similarities[max_similarity_idx]

    # Check if similarity exceeds threshold
    if max_similarity >= threshold:
        best_match = questions_df.iloc[max_similarity_idx]
        return {
            "question": best_match['question'],
            "answer": best_match['answer'],
            "similarity": max_similarity,
            "confidence": "high" if max_similarity >= 0.8 else "medium" if max_similarity >= 0.6 else "low"
        }
    else:
        return {
            "question": None,
            "answer": "I'm sorry, I couldn't find a good match for your question.",
            "similarity": max_similarity,
            "confidence": "low"
        }

# Test the model with some example queries
test_queries = [
    "What are the requirements for admission?",
    "How can I apply?",
    "Where is Zewail City?",
    "Do they offer PhD programs?",
    "What's the cost of tuition?",
    "Can I get financial assistance?",
    "Do they have dorms for students?",
    "I need to know about the GPA requirements",
    "Do international students need special documents?",
]

print("\nTesting the model with example queries:")
for query in test_queries:
    start_time = time.time()
    result = find_most_similar_question(query, df, transformer_model, encoder_model)
    end_time = time.time()

    print(f"\nQuery: {query}")
    print(f"Best match: {result['question']}")
    print(f"Answer: {result['answer'][:100]}...")
    print(f"Similarity: {result['similarity']:.4f}")
    print(f"Confidence: {result['confidence']}")
    print(f"Response time: {(end_time - start_time)*1000:.2f} ms")

```
