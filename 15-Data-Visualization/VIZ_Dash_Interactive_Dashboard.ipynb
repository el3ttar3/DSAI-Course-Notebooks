```python

data = pd.DataFrame({
    'Name': ['John', 'Alice', 'Bob', 'Carol'],
    'Age': [28, 24, 32, 29],
    'Salary': [60000, 55000, 70000, 62000]
})

app = dash.Dash(__name__)

app.layout = html.Div(
    [
        html.H1('Interactive Dash Dashboard'),
        dcc.Dropdown(
            id='data-selector',
            options=[
                {'label': 'Age', 'value': 'Age'},
                {'label': 'Salary', 'value': 'Salary'}
            ],
            value='Age'
        ),
        dcc.RadioItems(
            id='data-presentation',
            options=[
                {'label': 'Table', 'value': 'Table'},
                {'label': 'Chart', 'value': 'Chart'}
            ],
            value='Table',
            labelStyle={'display': 'block'}
        ),
        dcc.Graph(id='data-chart')
    ]
)

if __name__ == 'main':
    app.run_server(debug=True)
```
