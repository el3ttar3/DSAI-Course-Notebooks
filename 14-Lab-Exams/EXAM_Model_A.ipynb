```python
#Import the necessary modules:
import pandas as pd
import nltk
from nltk.stem import *
from nltk.stem.porter import *
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re # used to clean the data
```

```python
def clean(text):
   text = re.sub(r"http\S+", " ", text) # remove urls
   text = re.sub(r"RT ", " ", text) # remove rt
   text = re.sub(r"@[\w]*", " ", text) # remove handles
   text = re.sub(r"[\.\,\#_\|\:\?\?\/\=]", " ", text) # remove special characters
   text = re.sub(r'\t', ' ', text) # remove tabs
   text = re.sub(r'\n', ' ', text) # remove line jump
   text = re.sub(r"\s+", " ", text) # remove extra white space
   text = text.strip()
   return text
```

```python
# Get the list of English stopwords
stop_words = set(stopwords.words('english'))
```

```python
# Initialize Porter stemmer
stemmer = PorterStemmer()

def stemmed_text(data):
  results=[]
  tokens = word_tokenize(data)
  res = [stemmer.stem(tokens)]
  return res
```

```python
def remove_stop_words(data) :
  tokens = word_tokenize(data)
  fillterd_tokens=[tokee.lower() for tokee in tokens if tokee.lower() not in stop_words]
  return fillterd_tokens
```

```python
data = pd.read_csv("/content/amazon.csv")
# data.head(10)
data.drop(columns=["category"	,"discounted_price"	,"actual_price",	"discount_percentage"	,"rating"	,"rating_count","about_product",	"user_id"	,"user_name",	"review_id",	"review_title",	"img_link",	"product_link"],inplace= True)
data.head(10)
```

```python
def Preprocess(text):
  remove_stop_words(text)
  stemmed_text(text)
  return text
```

```python
data['product_name'].apply(clean)
data['review_content'].apply(clean)
data.head(10)
```

```python
data['product_name'].apply(Preprocess)
data['review_content'].apply(Preprocess)
data['docno']=data["product_id"].astype(str)
data.head(10)
```
