```python
import graphviz
from IPython.display import display

def display_current_flowchart():
    """
    Displays the current Order Fulfillment Process flowchart inline.
    Steps: Order Receipt -> Inventory Check -> Packing -> Shipping
    """
    dot = graphviz.Digraph(comment="Current Order Fulfillment Process")

    # Define the nodes
    dot.node('A', "Order Receipt\n(Customer places order)")
    dot.node('B', "Inventory Check\n(Manual verification)")
    dot.node('C', "Packing\n(Retrieve, pack, and label)")
    dot.node('D', "Shipping\n(Dispatch via logistics partners)")

    # Define the connections between nodes
    dot.edge('A', 'B')
    dot.edge('B', 'C')
    dot.edge('C', 'D')

    # Display the graph inline
    display(dot)

def display_redesigned_flowchart():
    """
    Displays the redesigned Order Fulfillment Process flowchart inline.
    Proposed changes: Automated Inventory Check and real-time dashboard.
    """
    dot = graphviz.Digraph(comment="Redesigned Order Fulfillment Process")

    # Define the nodes with new process elements
    dot.node('A', "Order Receipt\n(Customer places order)")
    dot.node('B', "Automated Inventory Check\n(ERP & barcode scanners)")
    dot.node('C', "Packing\n(Retrieve, pack, and label)")
    dot.node('D', "Shipping\n(Dispatch via logistics partners)")
    dot.node('E', "Real-time Logistics Dashboard\n(Track shipments)")

    # Define the process flow and additional links
    dot.edge('A', 'B')
    dot.edge('B', 'C')
    dot.edge('C', 'D')
    dot.edge('C', 'E', label="Update tracking", style="dashed")
    dot.edge('E', 'D', label="Feedback loop", style="dashed")

    # Display the graph inline
    display(dot)

# Display both flowcharts inline
display_current_flowchart()
display_redesigned_flowchart()
```
