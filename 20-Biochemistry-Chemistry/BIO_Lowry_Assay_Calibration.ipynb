```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Define standard concentrations and corresponding absorbance values
concentrations = np.array([0.0, 0.25, 0.50, 0.75, 1.0, 1.25])
absorbance = np.array([0.00, 0.18, 0.37, 0.55, 0.73, 0.92])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(concentrations, absorbance)

# Generate points for the best-fit line
x_fit = np.linspace(0, 1.3, 100)
y_fit = slope * x_fit + intercept

# Plot the data and regression line
plt.figure(figsize=(8, 6))
plt.scatter(concentrations, absorbance, color='blue', label='Standards')
plt.plot(x_fit, y_fit, color='red', label=f'Fit: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Protein Concentration (mg/ml)')
plt.ylabel('Absorbance (750 nm)')
plt.title('Calibration Curve for the Lowry Assay')
plt.legend()
plt.grid(True)
plt.show()

# Calculate unknown concentration
unknown_abs = 0.55
unknown_conc = (unknown_abs - intercept) / slope
print(f"Calculated protein concentration: {unknown_conc:.2f} mg/ml")
```
