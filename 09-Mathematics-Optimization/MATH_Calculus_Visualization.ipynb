```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 400)
y = np.abs(x)
plt.plot(x, y)
plt.axvline(0, color='red', linestyle= '--')
plt.title('NON-Differintible Fuction [X] ')
plt.show()
```

it has no slop, and not concave

```python
x = np.linspace(-2, 2, 400)
y = x**4 - x**2
plt.plot(x, y)
plt.title('Local Minima & saddle point')
plt.show()
```

-0.1, 0.7

```python
x = np.linspace(-10, 10, 400)
y = np.sin(x) + 0.1*x
plt.plot(x, y)
plt.title('Multi Modal Function')
plt.show()
```

-7 approx. is the global minima,

```python
import numpy as np
best = np.array([0.05, 50])
current = np.array([0.01, 20])
alpha = 0.1

noise = np.random.normal(0, 0.01, size=2)

new = current + alpha*(best - current) + noise

print('Old: ', current)
print('best: ', best)
print('New: ', new)
```
