```python
import math
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Point class to store coordinates in 3D space
class Point:
    def __init__(self, x, y, z):
        self.x = int(x)
        self.y = int(y)
        self.z = int(z)

# Generate all points in the 3D grid space
def generate_grid_points(grid_size=(10, 10, 10)):
    points = []
    for x in range(grid_size[0]):
        for y in range(grid_size[1]):
            for z in range(grid_size[2]):
                points.append(Point(x, y, z))
    return points

# Calculate Euclidean distance between two points in 3D space
def calculate_distance(p1, p2):
    return round(math.sqrt((p1.x - p2.x)**2 + (p1.y - p2.y)**2 + (p1.z - p2.z)**2), 2)

# Hill Climbing Algorithm to find the shortest path
def hill_climb(curr_node, next_node_list, dist_matrix):
    if not next_node_list:
        return None, 0

    min_cost = float('inf')
    next_node = None

    for node in next_node_list:
        if dist_matrix[curr_node][node] < min_cost:
            min_cost = dist_matrix[curr_node][node]
            next_node = node

    return next_node, min_cost

# Create a 3D plot for dynamic visualization
def plot_dynamic_path(points, selected_nodes, ax):
    path_x = [points[node].x for node in selected_nodes]
    path_y = [points[node].y for node in selected_nodes]
    path_z = [points[node].z for node in selected_nodes]

    ax.cla()
    ax.scatter([p.x for p in points], [p.y for p in points], [p.z for p in points], c='blue', marker='o')
    ax.plot(path_x, path_y, path_z, color='red', linewidth=2)  # Plotting the path
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.draw()

# Main function to run the Hill Climbing for drone navigation
def drone_navigation(grid_size=(10, 10, 10)):
    # Generate grid points
    points = generate_grid_points(grid_size)

    num = len(points)
    dist_matrix = [[0 for _ in range(num)] for _ in range(num)]

    # Calculate distance matrix between all points
    for i in range(num):
        for j in range(num):
            if i != j:
                dist_matrix[i][j] = calculate_distance(points[i], points[j])

    # Dynamic visualization setup
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Initialize the start node (0, 0, 0) and goal node (9, 9, 9)
    start_node = 0  # Starting point (0, 0, 0)
    goal_node = num - 1  # Goal point (9, 9, 9)

    remaining_nodes = list(range(num))
    selected_nodes = [start_node]
    remaining_nodes.remove(start_node)

    min_cost = 0

    # Animate the pathfinding process
    while remaining_nodes:
        curr_node = selected_nodes[-1]
        next_nodes = random.sample(remaining_nodes, k=max(1, len(remaining_nodes)//10))  # Sample a subset for next move

        next_node, cost = hill_climb(curr_node, next_nodes, dist_matrix)

        min_cost += cost
        remaining_nodes.remove(next_node)
        selected_nodes.append(next_node)

        # Update the path coordinates for visualization
        plot_dynamic_path(points, selected_nodes, ax)
        plt.pause(0.1)  # Pause to update the plot for dynamic visualization

        # Check if goal is reached
        if next_node == goal_node:
            break

    # Final Output
    print("Optimal Path:", selected_nodes)
    print("Minimum Cost:", min_cost)
    plt.show()

# Run the navigation
drone_navigation(grid_size=(10, 10, 10))
```
