```python
 import cv2
 import matplotlib.pyplot as plt
 import numpy as np
```

```python
img_1 = cv2.imread("/content/faces.jpeg")
plt.imshow(img_1)
```

```python
w = img_1.shape[1]
h = img_1.shape[0]
cx = w//3
cy = h//2
crops=[]

x = 0
y = 0
for i in range(2):
  x = 0
  for j in range(3):
    t = img_1[y:y+cy , x:x+cx]
    crops.append(t)
    x = x + cx
  y = y + cy
  fig= plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(6):

  plt.subplot(15, 3, i+1)
  plt.imshow(crops[i])
  plt.show()
```

```python
X_pairs = []
Y_labels = []
for i in range(len(crops)):

    for j in range(i+1, len(crops)):
        if i == j-1:
            label = 1
        else:
            label = 0
        face_pair = [crops[i], crops[j]]
        X_pairs.append(face_pair)
        Y_labels.append(label)
X = np.array(X_pairs)
Y = np.array(Y_labels)
print('X shape:', X.shape)
print('Y shape:', Y.shape)
```
