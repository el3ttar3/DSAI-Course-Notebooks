```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
# distance = math.sqrt((x2 - x1)^2 + (y2 - y1)^2 + (z2 - z1)^2)

P1 = np.array([24.13, -68.83, -68.83])
P2 = np.array([29.97, -73.41, -78.74])
P3 = np.array([47.50, -61.98, -72.90])
P4 = np.array([33.02, -77.22, -88.39])
P5 = np.array([-31.75, -22.61, -150.88])
P6 = np.array([30.23, -53.85, -53.85])

threshold = 13

neighbors = []
for point in [P1, P3, P4, P5, P6]:
    distance = np.sqrt((point[0]-P2[0])**2 + (point[1]-P2[1])**2 + (point[2]-P2[2])**2)
    if distance <= threshold:
        neighbors.append(point)

print("Neighbors of P2:", neighbors)
```

```python
voxel_size = 7

x_range = np.max([P1[0], P2[0], P3[0], P4[0], P5[0], P6[0]]) - np.min([P1[0], P2[0], P3[0], P4[0], P5[0], P6[0]])
y_range = np.max([P1[1], P2[1], P3[1], P4[1], P5[1], P6[1]]) - np.min([P1[1], P2[1], P3[1], P4[1], P5[1], P6[1]])
z_range = np.max([P1[2], P2[2], P3[2], P4[2], P5[2], P6[2]]) - np.min([P1[2], P2[2], P3[2], P4[2], P5[2], P6[2]])

num_voxels_x = int(x_range / voxel_size) + 1
num_voxels_y = int(y_range / voxel_size) + 1
num_voxels_z = int(z_range / voxel_size) + 1

print("Number of voxels in X plane:", num_voxels_x)
print("Number of voxels in Y plane:", num_voxels_y)
print("Number of voxels in Z plane:", num_voxels_z)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
for x in range(num_voxels_x):
    for y in range(num_voxels_y):
        for z in range(num_voxels_z):
            voxel_center = [
                np.min([P1[0], P2[0], P3[0], P4[0], P5[0], P6[0]]) + (x * voxel_size) + (voxel_size / 2),
                np.min([P1[1], P2[1], P3[1], P4[1], P5[1], P6[1]]) + (y * voxel_size) + (voxel_size / 2),
                np.min([P1[2], P2[2], P3[2], P4[2], P5[2], P6[2]]) + (z * voxel_size) + (voxel_size / 2)
            ]
            ax.scatter(voxel_center[0], voxel_center[1], voxel_center[2], color='blue')
```

```python
ax.scatter(P1[0], P1[1], P1[2], color='red', label='P1')
ax.scatter(P2[0], P2[1], P2[2], color='red', label='P2')
ax.scatter(P3[0], P3[1], P3[2], color='red', label='P3')
ax.scatter(P4[0], P4[1], P4[2], color='red', label='P4')
ax.scatter(P5[0], P5[1], P5[2], color='red', label='P5')
ax.scatter(P6[0], P6[1], P6[2], color='red', label='P6')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
