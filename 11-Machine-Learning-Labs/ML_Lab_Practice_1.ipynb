```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/nba.csv')
df.head(20)
```

```python
for i in range(len(df)):
  if pd.isna(df.loc[i]["College"]):
    df.at[i,'College']="UNK"
print(df)
```

```python
df['College'].fillna(df['College'].mode()[0], inplace= True)
df
```

```python
df['College'].fillna(df['College'].mode()[0], inplace= True)
df['Age'].fillna(df['Age'].mean(), inplace= True)
df['Age'].fillna(df['Age'].median(), inplace= True)
print(df)
```

```python
def replace_missing_with_mode(df, College):
  class_counts = df[College].value_counts()
  mode_value = class_counts.index[0]
  df[College].fillna(mode_value, inplace=True)
  replace_missing_with_mode(df, 'College')

print(df)
```

```python
sns.boxplot(x='Age', data=df)
plt.title('Player Age by Position')

plt.show()
```

```python
sns.scatterplot(data=df, x='Height', y='Weight')
plt.title('Height vs. Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
sns.barplot(x='Position', y='Age', data=df)
plt.xlabel('Position')
plt.ylabel('Age')
plt.show()
```

```python
import numpy as np

# Compute the correlation_matrix
correlation_matrix = df[['Age', 'Height', 'Weight', 'Salary']].corr()

# Create the heat_map
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Player Statistics')
plt.show()
```

```python
plt.figure()
sns.pairplot(data=df,hue="Weight")
plt.show()
```
