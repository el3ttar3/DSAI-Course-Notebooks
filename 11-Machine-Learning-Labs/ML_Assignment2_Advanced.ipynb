```python
# Function to remove stopwords
def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = []
    for word in tokens:
      if word.lower() not in stop_words:
        filtered_tokens.append(word.lower())

    print('Tokens are:',tokens,'\n')
    return ' '.join(filtered_tokens)
docs_df['job'] = docs_df['job'].astype(str)
docs_df['processed_text'] = docs_df['job'].apply(remove_stopwords)
```

```python
print('dataFrame after processing:\n')
docs_df
```

```python
from nltk.stem import *
from nltk.stem.porter import *
```

```python
# Initialize Porter stemmer
stemmer = PorterStemmer()
```

```python
def Steem_text(text):

    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    # print (tokens)
    return ' '.join(stemmed_tokens)
```

```python

docs_df['processed_text']=docs_df['processed_text'].apply(Steem_text)
docs_df.rename(columns={'Tweet_ID': 'docno'}, inplace=True)
docs_df
```

```python
indexer = pt.DFIndexer("./myFirstIndex", overwrite=True)
docs_df['docno'] = docs_df['job_ID'].astype(str)

# index the text, record the docnos as metadata
index_ref = indexer.index(docs_df['processed_text'], docs_df['docno'])
```

```python
print(index_ref.toString())

index = pt.IndexFactory.of(index_ref)

print(index.getCollectionStatistics().toString())
```

```python
for k in index.getLexicon():
  print("%s -> %s " % (k.getKey(), k.getValue().toString().split('@')[0]))
```

```python
for v in index.getLexicon():
  if 'ai' == v.getKey():
    print("%s --> %s " % (v.getKey(), v.getValue().toString().split('@')[0]))
```

```python
AI_Jops_IDs =[]
data = pt.BatchRetrieve(index_ref).search("ai")
for x in data:
  AI_Jops_IDs.append(data["docno"])
print(f"The job_IDs are:\n{AI_Jops_IDs[1]}")
```
