```python
# class_1
from abc import ABC, abstractmethod
class Account(ABC):
  def __init__(self):
    self._balance = 0.0
  def deposit(self, amount:float):
    if amount <= 0:
      raise ValueError("Deposit Amount Must Be Positive")
    self._balance += amount
  def get_balnce(self):
    return self._balance
  @abstractmethod
  def withdraw(self,amount:float):
    pass
```

```python
# class_2
class SavingsAccount(Account):
  def __init__(self, initial_balance = 0 , interest_rate = 0.05):
    super().__init__()
    self._balance = initial_balance
    self.interest_rate = interest_rate

  def calculate_interest(self):
    return self._balance * self.interest_rate

  def withdraw(self, amount):
    raise NotImplementedError("Withdrawals are not allowed from savings Acount")
```

```python
# class_3
class CheckingAccount(Account):
    def __init__(self, initial_balance=0):
        super().__init__()
        self._balance = initial_balance

    def withdraw(self, amount):
        if amount > self._balance:
            raise ValueError("Insufficient Balance")
        self._balance -= amount
```

```python
savings = SavingsAccount(1000)
savings.deposit(500)
print(savings.get_balnce())
print(savings.calculate_interest())
```

```python
checking = CheckingAccount(500)
checking.deposit(200)
checking.withdraw(400)
print(checking.get_balnce())

try:
    checking.withdraw(500)
except ValueError as e:
    print(e)
```

```python
savings = SavingsAccount(1000)
try:
    savings.withdraw(100)
except NotImplementedError as e:
    print(e)
```

```python
checking = CheckingAccount(500)
try:
    checking.deposit(-100)
except ValueError as e:
    print(e)
```
