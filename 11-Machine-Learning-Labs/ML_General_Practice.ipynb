```python
model_a = np.array([
    [-31.75, -22.61, -150.88],
    [-29.72, -33.02, -156.46],
    [62.48, -56.39, -88.14],
    [53.34, -66.55, -90.42],
    [20.07, 18.29, -33.53],
    [26.16, 31.50, -37.34]
])

model_b = np.array([
    [0.426768, 0.370185, -3.516970],
    [-0.218381, 0.888495, -3.504119],
    [-0.607195, 0.781828, 3.365147],
    [-1.765601, 0.427173, 3.189484],
    [-1.362596, 1.555820, 2.585355],
    [-1.361829, 1.432219, 2.883750],
    [-1.361241, 1.203836, 3.112131],
    [-1.822971, 1.021310, 1.755226],
    [-1.822638, 1.256445, 1.912337],
    [-1.822140, 1.413557, 2.147472]
])

centroid_a = np.mean(model_a, axis=0)
longest_dist_a = 0
longest_point_a = None
for i in range(len(model_a)):
    dist = ((model_a[i][0] - centroid_a[0])**2 + (model_a[i][1] - centroid_a[1])**2 + (model_a[i][2] - centroid_a[2])**2)**0.5
    if dist > longest_dist_a:
        longest_dist_a = model_a[i]
        longest_point_a = i
print("Longest point in Model A:", longest_dist_a)


centroid_b = np.mean(model_b, axis=0)
longest_dist_b = 0
longest_point_b = None
for i in range(len(model_b)):
    dist = ((model_b[i][0] - centroid_b[0])**2 + (model_b[i][1] - centroid_b[1])**2 + (model_b[i][2] - centroid_b[2])**2)**0.5
    if dist > longest_dist_b:
        longest_dist_b = dist
        longest_point_b = model_b[i]
print("Longest point in Model B:", longest_dist_b)
```

```python
dist_a = np.sqrt(np.sum((model_a - centroid_a)**2, axis=1))
dist_b = np.sqrt(np.sum((model_b - centroid_b)**2, axis=1))

points_a_03 = model_a[dist_a <= 0.3*longest_dist_a]
points_b_03 = model_b[dist_b <= 0.3*longest_dist_b]

print("Points in Model A with threshold 0.3:", points_a_03)
print("Points in Model B with threshold 0.3:", points_b_03)
```
