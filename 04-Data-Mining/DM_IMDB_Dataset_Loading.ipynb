```python
from kagglehub import dataset_load, KaggleDatasetAdapter
import os

dataset_dir = "/kaggle/input/imdb-dataset-of-50k-movie-reviews"
file_path = os.path.join(dataset_dir, "IMDB Dataset.csv")

# Correct call: adapter first, then path
df = dataset_load(
    KaggleDatasetAdapter.PANDAS,
    file_path
)

print(df.head())
```
