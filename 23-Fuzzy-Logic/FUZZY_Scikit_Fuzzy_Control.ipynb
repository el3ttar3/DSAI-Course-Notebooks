```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
# The_Inputs:
temp = ctrl.Antecedent(np.arange(0, 46, 1), 'Temperature')
humid = ctrl.Antecedent(np.arange(0, 101, 1), 'Humidity' )
```

```python
# The_Outputs:
speed = ctrl.Consequent(np.arange(0,101,1),'The Compressor Speed')
```

```python
temp['very low'] = fuzz.trimf(temp.universe, [0,0,15])
temp['low'] = fuzz.trimf(temp.universe, [0,15,30])
temp['high'] = fuzz.trimf(temp.universe, [15,30,45])
temp['very high'] = fuzz.trimf(temp.universe, [30,45,45])
```

```python
humid['dry'] = fuzz.trimf(humid.universe, [0,0,25])
humid['comfortable'] = fuzz.trimf(humid.universe, [0,25,50])
humid['humid'] = fuzz.trimf(humid.universe, [25,50,75])
humid['sticky'] = fuzz.trimf(humid.universe, [50,75,100])
```

```python
speed['off'] = fuzz.trimf(speed.universe, [0,0,25])
speed['low'] = fuzz.trimf(speed.universe, [0,25,50])
speed['medium'] = fuzz.trimf(speed.universe, [25,50,75])
speed['fast'] = fuzz.trimf(speed.universe, [50,75,100])
```

```python
# The_Rules:
rule1 = ctrl.Rule(temp['very low'] | humid['dry'], speed['off'])
rule2 = ctrl.Rule(temp['very low'] | humid['comfortable'], speed['off'])
rule3 = ctrl.Rule(temp['very low'] | humid['humid'], speed['off'])
rule4 = ctrl.Rule(temp['very low'] | humid['sticky'], speed['low'])
rule5 = ctrl.Rule(temp['low'] | humid['dry'], speed['off'])
rule6 = ctrl.Rule(temp['low'] | humid['comfortable'], speed['off'])
rule7 = ctrl.Rule(temp['low'] | humid['humid'], speed['low'])
rule8 = ctrl.Rule(temp['low'] | humid['sticky'], speed['medium'])
rule9  = ctrl.Rule(temp['high'] | humid['dry'], speed['low'])
rule10 = ctrl.Rule(temp['high'] | humid['comfortable'], speed['medium'])
rule11 = ctrl.Rule(temp['high'] | humid['humid'], speed['fast'])
rule12 = ctrl.Rule(temp['high'] | humid['sticky'], speed['fast'])
rule13 = ctrl.Rule(temp['very high'] | humid['dry'], speed['medium'])
```

```python
air_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13])
```

```python
air_simulator = ctrl.ControlSystemSimulation(air_ctrl)
```

```python
air_simulator.input['Temperature'] = 20
air_simulator.input['Humidity'] = 45

air_simulator.compute()
```

```python
print("The Air Conditioner Compressor Speed is at : " , round(air_simulator.output['The Compressor Speed']))
```
